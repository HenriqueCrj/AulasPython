```
import numpy as np
import turtle

#Aula 6 - Exercícios da TWiki - não funcionam completamente no Colab, então usei https://trinket.io/turtle
def exercise_6_1():
  def five_points_star():
    turt = turtle.Turtle()
    for i in range(5):
      turt.forward(100)
      turt.back(100)
      turt.left(72)
    turt.hideturtle()
  print("Questão 1:")
  five_points_star()

def exercise_6_2():
  def star(vertices,color,x,y):
    turt = turtle.Turtle()
    turt.hideturtle()
    turt.penup()
    turt.goto(x,y)
    turt.showturtle()
    turt.pendown()
    turt.color(color)
    for i in range(vertices):
      turt.forward(100)
      turt.back(100)
      turt.left(360/vertices)
    turt.hideturtle()
  print("Questão 2:")
  star(6,"blue",50,50)

def exercise_6_3():
  def flower(a,b,c,num):
    theta = np.linspace(0,2*np.pi,num)
    turt = turtle.Turtle()
    turt.hideturtle()
    turt.penup()
    r1 = a*np.cos(b*theta[0] + c)
    r2 = a*np.cos(2*b*theta[0] + c)
    r = r1 + r2
    turt.goto(r*np.cos(theta[0]),r*np.sin(theta[0]))
    turt.showturtle()
    turt.pendown()
    for angle in theta:
      r1 = a*np.cos(b*angle + c)
      r2 = a*np.cos(2*b*angle + c)
      r = r1 + r2
      turt.goto(r*np.cos(angle),r*np.sin(angle))
    turt.hideturtle()
  print("Questão 3:")
  flower(100,8,0,200)

def exercise_6_4():
  start = (0,80)
  turt = turtle.Turtle()
  def flower(a,b,c,num):
    theta = np.linspace(0,2*np.pi,num)
    turt.hideturtle()
    turt.penup()
    turt.color("red")
    r1 = a*np.cos(b*theta[0] + c)
    r2 = a*np.cos(2*b*theta[0] + c)
    r = r1 + r2
    turt.goto(start[0] + r*np.cos(theta[0]),start[1] + r*np.sin(theta[0]))
    turt.showturtle()
    turt.begin_fill()
    turt.pendown()
    for angle in theta:
      r1 = a*np.cos(b*angle + c)
      r2 = a*np.cos(2*b*angle + c)
      r = r1 + r2
      turt.goto(start[0] + r*np.cos(angle),start[1] + r*np.sin(angle))
    turt.end_fill()
    turt.hideturtle()
    print("Flower completed")

  def back_flower(a,b,c,d,num):
    theta = np.linspace(0,2*np.pi,num)
    turt.hideturtle()
    turt.penup()
    turt.color("blue")
    r = a*np.cos(b*theta[0] + c) + d
    turt.goto(start[0] + r*np.cos(theta[0]),start[1] + r*np.sin(theta[0]))
    turt.showturtle()
    turt.begin_fill()
    turt.pendown()
    for angle in theta:
      r = a*np.cos(b*angle + c) + d
      turt.goto(start[0] + r*np.cos(angle),start[1] + r*np.sin(angle))
    turt.end_fill()
    turt.hideturtle()
    print("Background flower completed")

  def stalk_and_leaf(a_s,b_s,c_s,num):
    def leaf(a_l,b_l,num):
      theta = np.linspace(-np.pi/2,0,num)
      turt.hideturtle()
      turt.penup()
      turt.color("green")
      turt.goto(start[0] + a_s ,start[1] - c_s + b_l + b_l*-1)
      turt.showturtle()
      turt.begin_fill()
      turt.pendown()
      for angle in theta:
        turt.goto(start[0] + a_s + a_l*np.cos(angle),start[1] - c_s + b_l + b_l*np.sin(angle))
      parameters = np.linspace(0,np.sqrt(a_l*a_l + b_l*b_l),num)
      for t in parameters:
        turt.goto(start[0] + a_l + a_s +(a_l*-t)/np.sqrt(a_l*a_l + b_l*b_l),start[1] - c_s + b_l + (b_l*(-t-1)/np.sqrt(a_l*a_l + b_l*b_l)))
      turt.end_fill()
      turt.hideturtle()
    theta = np.linspace(-np.pi/2,np.pi/2,num)
    turt.hideturtle()
    turt.penup()
    turt.color("brown")
    turt.goto(start[0],start[1])
    turt.showturtle()
    turt.pendown()
    for angle in theta:
      turt.goto(start[0] + a_s*np.cos(-angle) + b_s,start[1] + c_s*np.sin(-angle) - c_s)
    turt.hideturtle()
    leaf(100,40,num)
    print("Stalk and leaf completed")

  back_flower(30,7,0,60,170)
  flower(60,8,0,250)
  stalk_and_leaf(15,0,120,150)

exercise_6_1()
exercise_6_2()
exercise_6_3()
exercise_6_4()
```
